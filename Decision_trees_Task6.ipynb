# Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

# Dataset 

```python
iris = load_iris()
X = pd.DataFrame(iris.data[:, :], columns = iris.feature_names[:])
y = pd.DataFrame(iris.target, columns =["Species"])
```

# Splitting

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
first_tree = DecisionTreeClassifier(min_samples_split=4)
```

# Training

```python
first_tree.fit(X,y)
first_tree = first_tree.fit(X_train, y_train)
```

# Accuracy

```python
y_pred1 = first_tree.predict(X_test)
accuracy1 = metrics.accuracy_score(y_test, y_pred1)
accuracy1
```

# Tree Visualization

```python
Tree1 = plt.figure(figsize=(15,10))
firstVisualiztion = tree.plot_tree(first_tree, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
```
